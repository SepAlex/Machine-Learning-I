```python
import pandas as pd
import numpy as np

airline = pd.read_csv('https://raw.githubusercontent.com/wangrenfeng0/Machine-Learning-I/main/airline_passenger_satisfaction.csv') # read in the csv file

airline.head()
```

```python
airline.describe()
```

```python
print(airline.info())
print(airline.dtypes)
```

```python
if 'Unnamed: 0' in airline:
    del airline['Unnamed: 0']

print(airline.info())
```

```python
airline = airline.replace(to_replace='?',value=-1)
continuous_features = ['flight_distance','departure_delay_in_minutes']
airline[continuous_features] = airline[continuous_features].astype(np.float64)
airline.info()
airline.head()
airline.describe()
```

```python
airline = airline.replace(to_replace=-1,value=np.nan)
airline.info()
airline.describe()
airline.median()
```

```python
airline['age_range'] = pd.cut(airline.age,[0,16,65,120],3,labels=['child','adult','senior'])
airline.age_range.describe()

gender={'Male':1,'Female':0}
Customer_Type={'Loyal Customer':1,'disloyal Customer':0}
Type_of_Travel={'Personal Travel':1, 'Business travel':0}
Customer_Class = {'Eco':0,'Eco Plus':1,'Business':2}
Satisfaction ={'neutral or dissatisfied':0,'satisfied':1}
airline.Gender = [gender[item] for item in airline.Gender]
airline.customer_type = [Customer_Type[item] for item in airline.customer_type]
airline.type_of_travel = [Type_of_Travel[item] for item in airline.type_of_travel]
airline.customer_class = [Customer_Class[item] for item in airline.customer_class]
airline.satisfaction = [Satisfaction[item] for item in airline.satisfaction]
airline.head()
```

```python
airline_grouped = airline.groupby(by=['customer_type','age_range'])
print('Percentage of satisfaction rate in each customer type, separated by age groups:')
print(airline_grouped.satisfaction.sum()/airline_grouped.satisfaction.count()*100)
```

```python
airline_grouped = airline.groupby(by=['type_of_travel','age_range'])
print('Percentage of satisfaction rate in each travel type, separated by age groups:')
print(airline_grouped.satisfaction.sum()/airline_grouped.satisfaction.count()*100) #Business travel is 0, personal travel is 1
```

```python
airline_grouped = airline.groupby(by=['Gender','age_range'])
print('Percentage of satisfaction rate in each gender, separated by age groups:')
print(airline_grouped.satisfaction.sum()/airline_grouped.satisfaction.count()*100)
```

```python
airline_grouped = airline.groupby(by=['customer_class','age_range'])
print('Percentage of satisfaction rate in each class, separated by age groups:')
print(airline_grouped.satisfaction.sum()/airline_grouped.satisfaction.count()*100) #Satisfaction rate increases when class increases
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore', DeprecationWarning)
%matplotlib inline

satisfaction_counts = pd.crosstab([airline['customer_type'],airline['age_range']],airline.satisfaction.astype(bool))
satisfaction_rates = satisfaction_counts.div(satisfaction_counts.sum(1).astype(float),axis=0)
print(satisfaction_rates)
satisfaction_rates.plot(kind='barh',stacked=True,color=['black','gold'])
```

```python
satisfaction_counts = pd.crosstab([airline['type_of_travel'],airline['age_range']],airline.satisfaction.astype(bool))
satisfaction_rates = satisfaction_counts.div(satisfaction_counts.sum(1).astype(float),axis=0)
print(satisfaction_rates)
satisfaction_rates.plot(kind='barh',stacked=True,color=['black','gold'])
```

```python
vars_to_plot_separate = [['age'],
                         ['flight_distance'],
                         ['departure_delay_in_minutes'], ['arrival_delay_in_minutes']]

for index, plot_vars in enumerate(vars_to_plot_separate):
    plt.figure(figsize=(4, 4))
    ax = airline.boxplot(column=plot_vars, by='satisfaction')
plt.show()
```

```python
# make empirical CDFs
plt.figure(figsize=(20,10))
for fi,feature in enumerate(['age','flight_distance','departure_delay_in_minutes','arrival_delay_in_minutes']):
    plt.subplot(2,2,fi+1)
    x_ax = np.sort(airline[feature])
    y_ax = np.cumsum(x_ax)
    y_ax = y_ax/np.max(y_ax)
    plt.plot(x_ax,y_ax)
    plt.title(feature)
    
plt.show()
```

```python
# make percentile plots
plt.figure(figsize=(20,10))
for fi,feature in enumerate(['age','flight_distance','departure_delay_in_minutes','arrival_delay_in_minutes']):
    plt.subplot(2,2,fi+1)
    x_ax = np.linspace(0,100,10)
    y_ax = [np.percentile(airline[feature], x_val) for x_val in x_ax]
    plt.plot(x_ax,y_ax)
    plt.title(feature)
    
plt.show()
```
